# Análisis del impacto de campañas publicitarias

## Objetivo
Evaluar el desempeño de campañas publicitarias en distintos medios
utilizando métricas de alcance, interacción y retorno.

## Datos
Dataset en formato Excel con información estructurada de campañas.

## Metodología
- Carga y exploración de datos
- Cálculo de métricas promedio
- Generación de dashboard comparativo
- Exportación de reportes por área

## Resultados
Se observan diferencias relevantes en ROI y conversiones entre medios,
lo que permite apoyar decisiones estratégicas y operativas.

## Observaciones
El flujo desarrollado facilita la reutilización del análisis en otros proyectos.


```python
# Importación de librerías
import matplotlib.pyplot as plt

# Importación de funciones reutilizables
from funciones_analisis import (
    cargar_datos,
    calcular_metricas_por_medio,
    exportar_reportes_excel
)

# Carga del dataset
df = cargar_datos("campanas_publicidad.xlsx")

# Exploración inicial
df.head()

# Cálculo de métricas
resumen_medios = calcular_metricas_por_medio(df)

# Exportación de reportes para áreas de la empresa
exportar_reportes_excel(df, resumen_medios)

# Dashboard integrado
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.bar(resumen_medios.index, resumen_medios["roi"])
plt.title("ROI promedio por medio")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.bar(resumen_medios.index, resumen_medios["conversiones"])
plt.title("Conversiones promedio por medio")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```
